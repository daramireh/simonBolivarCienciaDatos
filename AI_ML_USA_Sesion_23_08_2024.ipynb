```python
# Paso 1: Definir las columnas numéricas y dicotómicas
numeric_columns = ['QualityRating', 'PriceRating', 'SalesImportance',
                   'ambianceRating', 'wifiRating', 'serviceRating', 'meetingRating']
categorical_columns = data.columns.difference(numeric_columns + ['Loyal'])

```

```python
# Paso 0: Revisar y eliminar columnas duplicadas
data = data.loc[:, ~data.columns.duplicated()]

# Paso 2: Eliminar columnas categóricas con más del 70% en una sola categoría
columns_to_drop_categorical = set()

for col in categorical_columns:
    # Calcula la proporción de la categoría más frecuente
    top_category_percentage = data[col].value_counts(normalize=True).max()
    if top_category_percentage > 0.7:
        print(f"Columna {col} tiene más del 70% en una sola categoría ({top_category_percentage*100:.2f}%) y será eliminada")
        columns_to_drop_categorical.add(col)

# Eliminar las columnas identificadas
data = data.drop(columns=columns_to_drop_categorical, inplace=False)

# Verificar las columnas eliminadas
print("Columnas eliminadas por tener más del 70% en una sola categoría:", columns_to_drop_categorical)
```

```python
from scipy.stats import chi2_contingency

# Eliminar columnas categóricas que cumplen el criterio
categorical_columns = categorical_columns.difference(columns_to_drop_categorical)

# Paso 3: Matriz de correlación de Spearman para las variables numéricas
numeric_correlation_matrix = data[numeric_columns].corr(method='spearman').abs()
upper_triangle_numeric = numeric_correlation_matrix.where(
    np.triu(np.ones(numeric_correlation_matrix.shape), k=1).astype(bool)
)

# Identificar las variables numéricas con alta correlación
columns_to_drop_numeric = [
    column for column in upper_triangle_numeric.columns if any(upper_triangle_numeric[column] > 0.6)
]

columns_to_drop_categorical = set()

# Paso 4: Prueba Chi-Cuadrado para las variables dicotómicas restantes
for i, col1 in enumerate(categorical_columns):
    if col1 not in columns_to_drop_categorical:  # Evitar columnas ya marcadas para eliminar
        for col2 in categorical_columns[i+1:]:
            if col2 not in columns_to_drop_categorical and col1 != col2:
                contingency_table = pd.crosstab(data[col1], data[col2])
                if contingency_table.shape[0] > 1 and contingency_table.shape[1] > 1:
                    _, p_value, _, _ = chi2_contingency(contingency_table)
                    if p_value < 0.05:  # Ajustar el umbral de significancia si es necesario
                        print(f"Alta asociación entre {col1} y {col2} (p-valor={p_value})")
                        columns_to_drop_categorical.add(col2)

# Paso 5: Eliminar las columnas seleccionadas
columns_to_drop = list(columns_to_drop_numeric) + list(columns_to_drop_categorical)
data = data.drop(columns=columns_to_drop, inplace=False)

# Verificar las columnas eliminadas y las restantes
print("Columnas eliminadas por multicolinealidad:", columns_to_drop)
print("Columnas restantes en el DataFrame:", data.columns)


```

```python
# Crear la fórmula con los nuevos nombres de columnas
formula = 'Loyal ~ ' + ' + '.join(data.columns.drop('Loyal'))

# Ajustar un modelo de regresión logística
model = smf.logit(formula, data=data)
fit = model.fit()

# Obtener un resumen del modelo
summary = fit.summary()
print(summary)
```

# Evaluando el Desempeño de Modelos de Clasificación

## 1. **Accuracy (Exactitud)**

**Definición:** Proporción de predicciones correctas sobre el total de predicciones.

**Fórmula:**

$$
\text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}}
$$

**Cuándo usarla:** Útil cuando las clases están equilibradas (es decir, similar número de ejemplos positivos y negativos).

---

## 2. **Precision (Precisión)**

**Definición:** Proporción de verdaderos positivos entre todos los ejemplos predichos como positivos.

**Fórmula:**

$$
\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
$$

**Cuándo usarla:** Importante cuando el costo de una predicción positiva incorrecta (FP) es alto, como en el diagnóstico médico.

---

## 3. **Recall (Sensibilidad o Tasa de Verdaderos Positivos)**

**Definición:** Proporción de verdaderos positivos detectados sobre el total de ejemplos positivos reales.

**Fórmula:**

$$
\text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}
$$

**Cuándo usarla:** Crítica cuando es más importante detectar todos los casos positivos, como en la detección de fraudes.

---

## 4. **F1-Score**

**Definición:** Media armónica entre precisión y recall, equilibra ambas métricas.

**Fórmula:**

$$
F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
$$

**Cuándo usarla:** Ideal cuando hay un desequilibrio entre las clases y es necesario un balance entre precisión y recall.

---

## 5. **AUC - ROC (Área Bajo la Curva ROC)**

**Definición:** Mide la capacidad del modelo para distinguir entre clases, representando la relación entre la tasa de verdaderos positivos y la tasa de falsos positivos.

**Interpretación:** Un AUC cercano a 1 indica un buen modelo; un AUC cercano a 0.5 indica un modelo que no distingue entre clases.

**Cuándo usarla:** Útil cuando se necesita evaluar el rendimiento de un modelo independientemente del umbral de decisión.

---

# **Conclusión:**

- **Selecciona la métrica adecuada:** Según el problema que estés resolviendo (balance entre clases, costo de errores).
- **Métricas complementarias:** A menudo es útil considerar múltiples métricas para obtener una imagen completa del rendimiento del modelo.

# Función de Perdida

## ¿Qué es la Función de Pérdida?
Imagina que eres un jugador de tiro al blanco. Tienes un arco y una flecha, y el objetivo es acertar en el centro del blanco. Cada vez que disparas la flecha, puedes medir qué tan lejos cayó del centro. La función de pérdida es como esa medida de cuán lejos estás del centro del blanco. Es una forma de calcular cuán incorrecto está tu modelo (tu disparo) al predecir un resultado.

### En machine learning:

Función de pérdida es una fórmula matemática que mide el error entre las predicciones de tu modelo y los valores reales. Por ejemplo, si estás prediciendo el precio de una casa, la función de pérdida podría medir la diferencia entre el precio predicho y el precio real.
Cuanto más cerca esté tu modelo de los valores reales, más baja será la pérdida. ¡Y el objetivo es hacer esa pérdida lo más pequeña posible!

# Gradiente descendiente

## ¿Qué es el Gradiente Descendente?
Volvamos al ejemplo del tiro al blanco. Supongamos que puedes ajustar la tensión de la cuerda del arco o la dirección de la flecha para mejorar tu puntería. El gradiente descendente es como un método que te dice en qué dirección ajustar para que tus disparos (predicciones) se acerquen cada vez más al centro del blanco (valor real).

### En machine learning:

Gradiente descendente es un algoritmo que ajusta los parámetros de tu modelo (como la dirección de tu flecha) para reducir la función de pérdida (es decir, para que el modelo cometa menos errores).
Funciona iterativamente: el algoritmo calcula en qué dirección debe mover los parámetros para reducir la pérdida y luego realiza ese ajuste. Lo hace una y otra vez hasta que la pérdida sea lo más pequeña posible.

## ¿Cómo se relacionan la Función de Pérdida y el Gradiente Descendente?
Imagina que estás en una colina y quieres llegar al punto más bajo (mínimo) de la colina. El punto más bajo es donde la pérdida es mínima, y el gradiente descendente es como una brújula que te dice en qué dirección bajar para llegar allí.

***Función de Pérdida:*** Es la "altura" de la colina en la que estás. Queremos encontrar el punto más bajo (mínima pérdida).

***Gradiente Descendente:*** Es el proceso de ir cuesta abajo (ajustar el modelo) para encontrar el punto más bajo en la colina (donde la función de pérdida es mínima).

# Modelos Boosting

¿Qué son los modelos Boosting?
Boosting es una técnica de ensamblaje utilizada en el aprendizaje automático para mejorar la precisión de los modelos de predicción. La idea principal detrás del boosting es combinar varios modelos débiles (modelos que apenas superan el azar) para crear un modelo fuerte y más preciso. A diferencia del bagging (como Random Forest), donde los modelos se entrenan en paralelo, en el boosting los modelos se entrenan de manera secuencial, donde cada modelo intenta corregir los errores cometidos por los modelos anteriores.

Principios clave del Boosting:

*   Modelos secuenciales: Los modelos se construyen uno tras otro, y cada nuevo modelo se enfoca en los errores cometidos por el modelo anterior.
*   Ajuste de pesos: En cada iteración, se ajustan los pesos de las observaciones, de modo que las observaciones mal predichas en una ronda tengan más importancia en la siguiente.
*   Combinación de modelos: Los modelos se combinan ponderadamente, donde modelos más recientes pueden tener más peso.



# XGBoost

¿Qué es XGBoost?
**XGBoost** (*Extreme Gradient Boosting*) es una implementación eficiente y optimizada del algoritmo de Gradient Boosting. Es ampliamente utilizado en la práctica debido a su rendimiento superior y su capacidad para manejar grandes conjuntos de datos y características complejas. XGBoost introduce varias mejoras sobre el boosting tradicional, como:

*   Optimización de memoria y velocidad: XGBoost está diseñado para ser rápido y utilizar eficientemente la memoria, lo que lo hace ideal para grandes volúmenes de datos.
*   Regularización: XGBoost incluye términos de regularización (L1 y L2) para prevenir el sobreajuste, lo que no está presente en el boosting tradicional.
*   Manejo de valores faltantes: XGBoost maneja de manera automática los valores faltantes durante el entrenamiento.
*   Paralelización: A diferencia de otros métodos de boosting, XGBoost puede paralelizar la construcción de árboles, lo que lo hace más rápido.


### Penalización L1 y L2:
Cuando entrenamos un modelo de machine learning, especialmente en problemas de clasificación o regresión, a veces queremos que el modelo no solo se ajuste bien a los datos, sino que también sea simple y evite aprender demasiado los detalles (lo que llamamos sobreajuste). Aquí es donde entran las penalizaciones L1 y L2.

#### Penalización L1 (Regularización Lasso)
Imagina que estás decorando una casa. Tienes un montón de cosas (variables o características) para elegir, pero no todas son necesarias. La penalización L1 es como un amigo estricto que te dice que solo puedes quedarte con las cosas más importantes y debes tirar las demás.
Matemáticamente, lo que hace la penalización L1 es empujar muchos de los coeficientes de tus variables a ser exactamente cero, lo que significa que el modelo las ignora por completo. Esto resulta en un modelo más sencillo que solo utiliza unas pocas características.

#### Penalización L2 (Regularización Ridge)
Ahora imagina que estás llenando una mochila para una caminata. Quieres llevar muchas cosas útiles, pero si llevas demasiado peso, la caminata será muy difícil. La penalización L2 es como un amigo que te dice que puedes llevar todo lo que quieras, pero que debes distribuir el peso uniformemente.
Matemáticamente, lo que hace la penalización L2 es reducir los coeficientes de las variables, pero sin llegar a empujarlos a cero como en L1. Esto significa que el modelo usa todas las características, pero no deja que ninguna tenga demasiado peso o influencia.

### ¿Qué es la Paralelización?
Imagina que tienes un gran número de tareas que hacer, como lavar los platos, barrer el piso, y preparar la cena. Si haces una tarea a la vez, te tomará mucho tiempo. Pero, si tienes varios amigos ayudándote y cada uno hace una tarea diferente al mismo tiempo, todo se hará mucho más rápido. ¡Eso es la paralelización!

Paralelización en Computación
En computación, la paralelización es el proceso de dividir un trabajo grande en tareas más pequeñas que pueden ser realizadas al mismo tiempo por diferentes "trabajadores" (en este caso, los procesadores o núcleos de la CPU). Esto hace que el trabajo se complete más rápido.
Ejemplo en Machine Learning: Cuando entrenamos modelos complejos como XGBoost, podemos usar la paralelización para entrenar diferentes partes del modelo al mismo tiempo en lugar de una por una. Esto acelera significativamente el tiempo de entrenamiento, especialmente cuando trabajamos con grandes conjuntos de datos.

# Support Vector Machine

## ¿Qué es el Modelo Support Vector Machine (SVM)?
**Support Vector Machine** (*SVM*) es un algoritmo de aprendizaje supervisado utilizado principalmente para problemas de clasificación, aunque también puede ser usado para regresión. La idea principal detrás de SVM es encontrar la línea o el hiperplano que mejor separa las clases en un espacio de características.

### ¿Cómo funciona SVM?
Imagina que tienes dos tipos de frutas, manzanas y naranjas, y quieres separarlas en función de sus características, como peso y tamaño. SVM encuentra la línea (o el hiperplano en dimensiones más altas) que divide estos dos grupos de frutas de manera que las frutas estén lo más separadas posible.

***Separación Máxima:*** SVM busca la línea que no solo separa las dos clases, sino que lo hace con el mayor margen posible. Esto significa que SVM intenta maximizar la distancia entre las frutas más cercanas a la línea de separación en ambas clases.

***Vectores de Soporte:*** Las frutas que están más cercanas a la línea de separación se llaman vectores de soporte. Estos son los puntos clave que SVM utiliza para definir la línea de separación.

***Hiperplano:*** Si tienes más de dos características, SVM utiliza un hiperplano (una "línea" en un espacio de dimensiones superiores) para separar las clases.
